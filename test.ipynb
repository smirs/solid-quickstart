```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame({'year':['2023', '2024', '2025', '2026', '2027', '2028', '2029', '2030'], 
                   'num_prop':[1,2,3,4,5,6,7,8],
                   'req_fund':[150, 135, 120, 105, 90, 75, 60, 45]
                   })
df['exp_inco'] = df['num_prop']*1500*12
df
```

```python
sns.barplot(x='year', y='exp_inco', color='#c8c045', data=df)
```

```python
sns.barplot(x='year', y='num_prop', color='#94c045', data=df)
```

```python
sns.barplot(x='year', y='req_fund', color='#94d6a5', data=df)
```
